# (c) Ajouter le code suivant pour nettoyer les donnГ©es et supprimer les stopwords.

```python
import re
import unicodedata
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
#nltk.download()
stop_words = stopwords.words('english')

# Converts the unicode file to ascii
def unicode_to_ascii(s):
    return ''.join(c for c in unicodedata.normalize('NFD', s)
        if unicodedata.category(c) != 'Mn')

def preprocess_sentence(w):
    w = unicode_to_ascii(w.lower().strip())
    # creating a space between a word and the punctuation following it
    # eg: "he is a boy." => "he is a boy ."
    w = re.sub(r"([?.!,Вҝ])", r" \1 ", w)
    w = re.sub(r'[" "]+', " ", w)
    # replacing everything with space except (a-z, A-Z, ".", "?", "!", ",")
    w = re.sub(r"[^a-zA-Z?.!]+", " ", w)
    w = re.sub(r'\b\w{0,2}\b', '', w)

    # remove stopword
    mots = word_tokenize(w.strip())
    mots = [mot for mot in mots if mot not in stop_words]
    return ' '.join(mots).strip()

df.review = df.review.apply(lambda x :preprocess_sentence(x))
df.head(), stop_words
```

## ModГ©lisation
Nous allons implГ©menter un modГЁle Word2Vec particulier : le modГЁle Continuous Bag Of Words (CBOW). Le modГЁle CBOW cherche Г  prГ©dire un mot grГўce Г  son contexte, c'est-Г -dire grГўce aux mots proches de lui dans le texte. Les inputs du modГЁle sont les mots du contexte et l'output du modГЁle est une probabilitГ© de prГ©diction du mot cible.

Plus prГ©cisemment, le modГЁle CBOW est un rГ©seau de neurones Г  3 couches : une couche d'input, une couche cachГ©e et une couche ouput. La couche cachГ©e est constituГ©e d'une couche Embedding qui transforme chaque mot input en un vecteur d'embedding, de tel sorte que la matrice d'embedding est apprise au fur et Г  mesure de l'entrainement. Il y a aussi une couche de Pooling (GlobalAveragePooling1D) qui somme les diffГ©rents embeddings pour obtenir un rГ©sultat de bonne dimension. Enfin la prГ©diction du mot cible est faite grГўce Г  une couche Dense.

(g) Ajouter le code suivant pour crГ©er l'ensemble de donnГ©es (X, Y).

```python
import numpy as np


def sentenceToData(tokens, WINDOW_SIZE):
    window = np.concatenate((np.arange(-WINDOW_SIZE,0),np.arange(1,WINDOW_SIZE+1)))
    X,Y=([],[])
    for word_index, word in enumerate(tokens) :
        if ((word_index - WINDOW_SIZE >= 0) and (word_index + WINDOW_SIZE <= len(tokens) - 1)) :
            X.append(word)
            Y.append([tokens[word_index-i] for i in window])
    return X, Y


WINDOW_SIZE = 5

X, Y = ([], [])
for review in df.review:
    for sentence in review.split("."):
        word_list = tokenizer.texts_to_sequences([sentence])[0]
        if len(word_list) >= WINDOW_SIZE:
            Y1, X1 = sentenceToData(word_list, WINDOW_SIZE//2)
            X.extend(X1)
            Y.extend(Y1)
    
X = np.array(X).astype(int)
y = np.array(Y).astype(int).reshape([-1,1])
```

```python
import numpy as np
import tensorflow as tf
from  pathlib import Path

def test_model_load(mdlPath):
    # 1. Charger
    mdlPath=Path(mdlPath)
    mdl = tf.keras.models.load_model(mdlPath)
    assert mdl is not None, "ModГЁle non chargГ©"

    # 2. VГ©rifier l'architecture
    print(mdl.summary())

    # 3. VГ©rifier les poids non nuls
    weights = mdl.get_weights()
    assert len(weights) > 0, "Aucun poids chargГ©"
    assert any(np.any(w != 0) for w in weights), "Tous les poids sont nuls !"

    # 4. Tester une prГ©diction avec un input factice
    input_shape = mdl.input_shape[1:]  # ex: (224, 224, 3)
    dummy = np.zeros((1, *input_shape))
    output = mdl.predict(dummy)

    # 5. VГ©rifier la forme de la sortie
    assert output.shape[0] == 1, "Sortie inattendue"
    print(f"Sortie OK : shape={output.shape}, valeurs={output}")

test_model_load('data/word2vec.keras')
```

La similitude est une mГ©trique mesurant la distance entre deux mots. Cette distance reprГ©sente la faГ§on dont les mots sont liГ©s entre eux.

(k) Ajouter le code suivant pour extraire la matrice d'embeddings et dГ©finir les fonctions de similitude.

```python
import numpy as np
from sklearn.preprocessing import Normalizer

vectors = model.layers[0].trainable_weights[0].numpy()


def dot_product(vec1, vec2):
    return np.sum((vec1*vec2))

def cosine_similarity(vec1, vec2):
    return dot_product(vec1, vec2)/np.sqrt(dot_product(vec1, vec1)*dot_product(vec2, vec2))

def find_closest(word_index, vectors, number_closest):
    list1=[]
    query_vector = vectors[word_index]
    for index, vector in enumerate(vectors):
        if not np.array_equal(vector, query_vector):
            dist = cosine_similarity(vector, query_vector)
            list1.append([dist,index])
    return np.asarray(sorted(list1,reverse=True)[:number_closest])

def compare(index_word1, index_word2, index_word3, vectors, number_closest):
    list1=[]
    query_vector = vectors[index_word1] - vectors[index_word2] + vectors[index_word3]
    normalizer = Normalizer()
    query_vector =  normalizer.fit_transform([query_vector], 'l2')
    query_vector= query_vector[0]
    for index, vector in enumerate(vectors):
        if not np.array_equal(vector, query_vector):
            dist = cosine_similarity(vector, query_vector)
            list1.append([dist,index])
    return np.asarray(sorted(list1,reverse=True)[:number_closest])

def print_closest(word, number=10):
    index_closest_words = find_closest(word2idx[word], vectors, number)
    for index_word in index_closest_words :
        print(idx2word[index_word[1]]," -- ",index_word[0])
```
